# 1. SQL Data Ingestion (Raw Text)

This notebook loads the RAW EMR data into PostgreSQL for Vanna to query. We no longer use manual HDBSCAN clustering.

```python
import os
import sys
import pandas as pd
from sqlalchemy import create_engine, text

cwd = os.path.abspath(os.getcwd())
project_root = None

temp = cwd
for _ in range(4):
    if os.path.exists(os.path.join(temp, 'src', 'config.py')):
        project_root = temp
        break
    parent = os.path.abspath(os.path.join(temp, '..'))
    if parent == temp:
        break
    temp = parent

if not project_root:
    for path in list(sys.path):
        if not path:
            continue
        candidate = os.path.abspath(path)
        for folder in [candidate, os.path.abspath(os.path.join(candidate, '..')), os.path.abspath(os.path.join(candidate, '..', '..'))]:
            if os.path.exists(os.path.join(folder, 'src', 'config.py')):
                project_root = folder
                break
        if project_root:
            break

if project_root:
    if project_root not in sys.path:
        sys.path.insert(0, project_root)
    path_prefix = os.path.relpath(project_root, cwd)
    print(f"Project root found: {project_root}")
    print(f"Current working directory: {cwd}")
    print(f"Relative path prefix: {path_prefix}")
else:
    print("Warning: Could not automatically detect project root. Using default path prefix '..'")
    path_prefix = ".."

from src.config import settings
```

```python
print("Setup: Membuat tabel site_reference dari plotingSite.csv...")
engine_setup = create_engine(settings.postgres_url)

with engine_setup.connect() as conn:
    conn.execute(text("COMMIT"))
    conn.execute(text("""
        CREATE TABLE IF NOT EXISTS site_reference (
            code VARCHAR(10) PRIMARY KEY,
            full_name VARCHAR(100) NOT NULL
        )
    """))
    conn.commit()
    existing = conn.execute(text("SELECT COUNT(*) FROM site_reference")).scalar()

if existing == 0:
    csv_path = os.path.join(path_prefix, "data", "plottingSite.csv")
    df_sites = pd.read_csv(csv_path, header=None, names=["code", "full_name"], encoding="latin1")
    df_sites["code"] = df_sites["code"].str.strip()
    df_sites["full_name"] = df_sites["full_name"].str.strip()
    df_sites.to_sql("site_reference", engine_setup, if_exists="append", index=False)
    total = len(df_sites)
    print(f"  Tabel site_reference dibuat dengan {total} baris.")
else:
    print(f"  Tabel site_reference sudah ada ({existing} baris), skip insert.")
```

```python
print("Setup: Membuat tabel account_reference dari data/account_lookup.csv...")
engine_setup = create_engine(settings.postgres_url)

with engine_setup.connect() as conn:
    conn.execute(text("COMMIT"))
    conn.execute(text("""
        CREATE TABLE IF NOT EXISTS account_reference (
            full_name VARCHAR(200) PRIMARY KEY
        )
    """))
    conn.commit()
    existing = conn.execute(text("SELECT COUNT(*) FROM account_reference")).scalar()

if existing == 0:
    csv_path = os.path.join(path_prefix, "data", "account_lookup.csv")
    df_accts = pd.read_csv(csv_path, encoding="utf-8")
    df_accts.to_sql("account_reference", engine_setup, if_exists="append", index=False)
    total = len(df_accts)
    print(f"  Tabel account_reference dibuat dengan {total} baris.")
else:
    print(f"  Tabel account_reference sudah ada ({existing} baris), skip insert.")
```

```python
print("1. Loading Raw Data...")
file_path = os.path.join(path_prefix, "data", "Dashboard EMR(report1776669858353).csv")
if not os.path.exists(file_path):
    raise FileNotFoundError("Dashboard EMR CSV not found in data folder.")

try:
    df = pd.read_csv(file_path, encoding='utf-8-sig')
except UnicodeDecodeError:
    print("  UTF-8 decoding failed — falling back to 'latin1' encoding")
    df = pd.read_csv(file_path, encoding='latin1')
except Exception as e:
    print(f"  Fallback reading using cp1252 due to {e}")
    df = pd.read_csv(file_path, encoding='cp1252')

print(f"Loaded {len(df)} records.")
```

```python
print("2. Cleaning and Formatting Data for SQL...")

df.columns = [
    c.strip().lower().replace(' / ', '_').replace(' ', '_').replace('/', '_').replace('(', '').replace(')', '').replace('?', '')
    for c in df.columns
]

if 'model_family' not in df.columns and 'machine_model' in df.columns:
    print("  Creating 'model_family' column from 'machine_model'...")
    df['model_family'] = df['machine_model'].apply(lambda x: str(x).split('-')[0] if pd.notna(x) else x)

date_cols = ['created_date', 'emr_last_closed_date']
for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')

df = df.loc[:, ~df.columns.duplicated()]

print("Columns ready for SQL:", df.columns.tolist())
```

```python
print("3. Connecting to PostgreSQL and Ingesting Data...")
engine = create_engine(settings.postgres_url)

table_name = 'emr_records'
print(f"Writing to table '{table_name}'...")

df.to_sql(table_name, engine, if_exists='replace', index=False)
print("Ingestion complete.")
```

```python
print("4. Testing SQL Query...")
test_query = f"SELECT machine_model, COUNT(*) as fault_count FROM {table_name} GROUP BY machine_model ORDER BY fault_count DESC LIMIT 5"
result = pd.read_sql(test_query, engine)
print(result)
```

```python
print("5. Backfilling techcare_component from duplicate column...")
with engine.begin() as conn:
    result = conn.execute(
        text("""
            UPDATE emr_records 
            SET techcare_component = "techcare_component.1"
            WHERE techcare_component IS NULL AND "techcare_component.1" IS NOT NULL
        """)
    )
    print(f"  Updated {result.rowcount} rows")
    
    result2 = conn.execute(
        text("""
            UPDATE emr_records 
            SET part_description = "part_description.1"
            WHERE part_description IS NULL AND "part_description.1" IS NOT NULL
        """)
    )
    print(f"  Updated part_description {result2.rowcount} rows")
    
    row = conn.execute(
        text("SELECT COUNT(*), COUNT(techcare_component), COUNT(*) - COUNT(techcare_component) FROM emr_records")
    ).fetchone()
    print(f"  Total: {row[0]}, Non-NULL: {row[1]}, Still NULL: {row[2]}")
    
    conn.execute(text('ALTER TABLE emr_records DROP COLUMN IF EXISTS "techcare_component.1"'))
    conn.execute(text('ALTER TABLE emr_records DROP COLUMN IF EXISTS "part_description.1"'))
    print("  Redundant columns dropped.")
```

```python
# =================================================================
# Step 6: Run schema migrations (idempotent -- safe to re-run)
# Adds columns: smr_trouble, ppi_record_id, ppi_external_id, etc.
# These columns are NOT in the raw CSV, so they must be added via
# migration after every df.to_sql(..., if_exists='replace') call.
# =================================================================
print('Running schema migrations...')

try:
    from scripts.migrate_smr import run_migration as run_smr_migration
    run_smr_migration()
    print('  [OK] migrate_smr done (smr_trouble column)')
except Exception as e:
    print(f'  [WARN] migrate_smr failed: {e}')

try:
    from scripts.migrate_ppi import run_migration as run_ppi_migration
    run_ppi_migration()
    print('  [OK] migrate_ppi done (ppi_record_id, ppi_external_id, ...)')
except Exception as e:
    print(f'  [WARN] migrate_ppi failed: {e}')

print('All migrations complete.')
```

> **Catatan penting:** Migration ini idempotent (aman dijalankan berkali-kali).
> Kolom `smr_trouble`, `ppi_record_id`, dan `ppi_*` tidak ada di raw CSV -- mereka ditambahkan via migration.
> Setiap kali kamu re-run Step 3 (`df.to_sql(..., if_exists='replace')`), tabel akan di-DROP dan dibuat ulang tanpa kolom tersebut.
> **Jadi Step 6 ini wajib dijalankan setelah Step 3.**
