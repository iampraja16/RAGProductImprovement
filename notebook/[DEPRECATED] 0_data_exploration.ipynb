# 0. Data Exploration

This notebook is for exploring the raw EMR (Equipment Maintenance Record) data from the Excel file before processing.

```python
import os
import sys
import pandas as pd

# Add src to path
sys.path.append(os.path.abspath('..'))
from src.config import settings
```

```python
# Load the data
file_path = os.path.join("..", settings.data_dir, settings.emr_file_name)
print(f"Loading data from {file_path}...")
df = pd.read_excel(file_path, sheet_name=settings.emr_sheet_name)
print(f"Loaded {len(df)} rows and {len(df.columns)} columns.")
df.head()
```

```python
# Check missing values
df.isnull().sum()
```

```python
# Distribution of Machine Models
df['Machine Model'].value_counts().head(10).plot(kind='bar', title='Top 10 Machine Models')
```

```python
# Distribution of Sites
df['Branch / Site'].value_counts().head(10).plot(kind='bar', title='Top 10 Sites')
```

```python
# Preview unstructured text fields
text_cols = ['Subjects', 'Symptom', 'Caused of Problem']
df[text_cols].dropna().head()
```
